```python
Q1 = florida['TransactionAmount'].quantile(0.25)
Q3 = florida['TransactionAmount'].quantile(0.75)
IQR = Q3 - Q1

lower_limit =0
upper_limit = Q3 + 1.5 * IQR    
outliers = florida[(florida['TransactionAmount'] < lower_limit) | (florida['TransactionAmount'] > upper_limit)]
outliers
```

```python
#2• Hər bir əməliyyatın müddətini əks etdirən yeni sütun yaradın (TransactionEndDatetime - TransactionStartDatetime)
florida["TransactionStartDateTime"]=pd.to_datetime(florida["TransactionStartDateTime"],format='%m/%d/%Y %H:%M')
florida["TransactionEndDateTime"]=pd.to_datetime(florida["TransactionEndDateTime"],format='%m/%d/%Y %H:%M')
```

```python
florida['TransactionDuration'] = (florida['TransactionEndDateTime'] - florida['TransactionStartDateTime']).dt.total_seconds()
```

```python
#5• Qadınlar və kişilərin tranzaksiya məbləği arasında gözə çarpan ferqin olub olmamasini hipotez testi vasitəsilə araşdırın
from scipy.stats import ttest_ind
merged_gender_df = pd.merge(florida, customers[['CardholderID', 'Gender']], on='CardholderID', how='left')
female_amounts = merged_gender_df[merged_gender_df['Gender'] == 'F']['TransactionAmount']
male_amounts = merged_gender_df[merged_gender_df['Gender'] == 'M']['TransactionAmount']
t_stat, p_value = ttest_ind(male_amounts, female_amounts, equal_var=False)
if p_value < 0.05:
    print("Nəticə: Qadınlar və kişilərin tranzaksiya məbləğləri arasında əhəmiyyətli fərq var.")
else:
    print("Nəticə: Əhəmiyyətli fərq tapılmadı.")
```

```python
print("Male count:", len(male_amounts))
print("Female count:", len(female_amounts))
```

```python
#7• Tranzaksiya məbləğlərini vizuallaşdırmaq üçün bar diaqramları və ya histoqramlar yaradın
plt.figure(figsize=(10, 6))
sns.histplot(florida['TransactionAmount'], bins=30, kde=True)
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
#8• Tranzaksiya tipləri ilə bağlı pie diaqramı yaradın.
```

```python
merged_df=pd.merge(florida,transaction,on="TransactionTypeID",how="left")
transaction_type_counts=merged_df["TransactionTypeName"].value_counts()
plt.figure(figsize=(8,8))
transaction_type_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3'))
plt.title("Transaction Type Distribution")
plt.show()
```

```python
merged_df = pd.merge(florida, customers[['CardholderID', 'ATMID']], on='CardholderID', how='left')
atm_counts = customers.groupby('ATMID').size().reset_index(name='No of ATMs')
merged_atm_df = pd.merge(merged_df, atm_counts, on='ATMID', how='left')
plt.figure(figsize=(10, 6))
sns.scatterplot(x='No of ATMs', y='TransactionAmount', data=merged_atm_df)
plt.title('Relationship between Number of ATMs and Transaction Amount')
plt.xlabel('Number of ATMs')
plt.ylabel('Transaction Amount')
plt.show()
```
